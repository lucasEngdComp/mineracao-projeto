```python
import pandas as pd
import numpy as np
```

```python
dados1 = pd.read_csv('crimes.csv', encoding='latin-1', sep=';')
dados2 = pd.read_csv('homicidio.csv', encoding='latin-1', sep=';')
dados3 = pd.read_csv('veiculos_roubados.csv', encoding='latin-1', sep=';')
dados4 = pd.read_csv('furtos_consumado.csv', encoding='latin-1', sep=';')

df_concatenado = pd.concat([dados1,dados2,dados3,dados4], ignore_index=True)
df_concatenado
```

```python
#Removendo os valores NaN
data_sem_nan = df_concatenado.fillna(0)
data_sem_nan
```

```python
# dados repetidos ?
dado_repetido = data_sem_nan.duplicated()

# Exiba as linhas duplicadas (se houver)
df_duplicatas = data_sem_nan[dado_repetido]
print(df_duplicatas)
 
```
